```python
import json
import ifcopenshell
import ifcHelper 
import pandas as pd
import numpy as np
import lcabygjson
import uuid
```

```python
file_name =  'exceldata.xls'
sheet_products =  'Products' # sheet name or sheet number or list of sheet numbers and names

try:
  products_df = pd.read_excel(io=file_name, sheet_name=sheet_products)
  print("products Excel File loaded successfully")
  print(products_df.head(1))  # print first 5 rows of the dataframe
except:
  print("An exception occurred")
```

```python
productids = products_df['Productid']
product_names = products_df['Row Labels']
amount = products_df['Amount']
stageA1A3_id = products_df['A1A3']
stageC3_id = products_df['C3']
stageC4_id = products_df['C4']
stageD_id = products_df['D']
```

```python
json_array_products = []
```

```python
i=0
products_df
for id in productids:
    comment = 'Material created from Excel Import ' 
    print (i)
    product_properties = {'id': productids[i],  'name': {'Danish': product_names[i], 'English': product_names[i]}, 'source': 'User','comment': comment,'locked': True}
    product_set = {'Product': product_properties}
    product_node = {'Node': product_set}
    json_array_products.append(product_node)
    
    product_edge_details = {'id': str(uuid.uuid4()), 'enabled': True}
    product_edge_data = [{'ProductToStage': product_edge_details}, productids[i], stageA1A3_id[i]]
    product_edge = {'Edge': product_edge_data}
    json_array_products.append(product_edge)
    
    current_id = stageC3_id[i]
    if str(current_id) == 'nan' or str(current_id) == 'XXX':
        print(i)
        print('Phase empty')
    else:
        product_edge_details = {'id': str(uuid.uuid4()), 'enabled': True}
        product_edge_data = [{'ProductToStage': product_edge_details}, productids[i], stageC3_id[i]]
        product_edge = {'Edge': product_edge_data}
        json_array_products.append(product_edge)
    
    current_id = stageC4_id[i]
    if str(current_id) == 'nan' or str(current_id) == 'XXX':
        print(i)
        print('Phase empty')
    else:
        product_edge_details = {'id': str(uuid.uuid4()), 'enabled': True}
        product_edge_data = [{'ProductToStage': product_edge_details}, productids[i], stageC4_id[i]]
        product_edge = {'Edge': product_edge_data}
        json_array_products.append(product_edge)
    
    current_id = stageD_id[i]
    if str(current_id) == 'nan' or str(current_id) == 'XXX':
        print(i)
        print('Phase empty')
    else:
        product_edge_details = {'id': str(uuid.uuid4()), 'enabled': True}
        product_edge_data = [{'ProductToStage': product_edge_details}, productids[i], stageD_id[i]]
        product_edge = {'Edge': product_edge_data}
        json_array_products.append(product_edge)
    
    
    
    i += 1
```

```python
json_array_products
```

```python
try:
  with open('dk-lcabyg_json/products.json', 'w') as outfile:
        json.dump(json_array_products, outfile,indent=4)

  print("Files exported successfully")
except:
  print("An exception occurred")
```
